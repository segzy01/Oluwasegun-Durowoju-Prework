```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import numpy
import keras
```

#Data Preparation

```python
# define the raw data
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
# set the sequence length
seq_length = 3 #2, 3, 5
# create a dictionary(map) of the letter index to character
char_to_int = dict((c, i) for i, c in enumerate(alphabet))
int_to_char = dict((i, c) for i, c in enumerate(alphabet))
dataX = []
dataY = []
for i in range(0, len(alphabet) - seq_length):
    seq_in = alphabet[i:i + seq_length]
    seq_out = alphabet[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
    print(seq_in, '->', seq_out)

# now dataX and dataY can be used for training a neural network or other machine learning model
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
seq_length = 3 # MAX time steps
X_dim = 1 
# define the size of the hidden layer
hidden_size = 32
```

```python
# create the LSTM model
model = Sequential()
model.add(LSTM(hidden_size, input_shape=(seq_length, X_dim)))
model.add(Dense(len(alphabet), activation='softmax'))
model.compile(loss='categorical_crossentropy', metrics='accuracy', optimizer='adam')
```

```python
# prepare the data for training
X = numpy.reshape(dataX, (len(dataX), seq_length, 1))
Y = keras.utils.to_categorical(dataY)
```

```python
#YPred = model.predict(X)
#print(YPred)
```

```python
# train the model
model.fit(X, Y, epochs=500, batch_size=32) #32, 64, 72,
```

```python
# generate a sequence of characters
start_seq = 'ABC'
seq = start_seq
for i in range(100):
    x = numpy.reshape([char_to_int[char] for char in seq], (1, seq_length, 1))
    prediction = model.predict(x, verbose=0)
    index = numpy.argmax(prediction)
    result = int_to_char[index]
    seq = seq[1:] + result
    print(result, end='')
```

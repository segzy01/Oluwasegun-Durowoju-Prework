##1. Using Python `requests` library, retrieve the WA state electric vehicle population data set from data.gov

Data.gov dataset: https://catalog.data.gov/dataset/electric-vehicle-population-data

Requests Docs: https://requests.readthedocs.io/en/latest/

```python
import requests
url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD'

### START YOUR CODE HERE ###
# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Save the response content to a file
    with open('electric_vehicle_population.csv', 'wb') as file:
        file.write(response.content)
    print("Data downloaded successfully.")
else:
    print("Failed to retrieve the data. Check the URL or try again later.")
```
